```python
#TRAINING DATA
"""
- may need to run hyperopt on full dataset rather than per neuron's trials
"""
try:
    #initialize group(s)
    groups = [
        'CueRein-LCDD',
        'CueRein-DREADDs',

        # 'CueReinCNO-LCDD',
        # 'CueReinCNO-DREADDs',

        # 'CueReinPersist-LCDD',
        # 'CueReinPersist-DREADDs',

        'CueRein-All'
        ]
    indir_list = [
        LCDD_cueRein_dir,
        Gi_cueRein_dir,

        # LCDD_cueReinCNO_dir,
        # Gi_cueReinCNO_dir,

        # LCDD_cueReinPersist_dir,
        # Gi_cueReinPersist_dir,

        ALL_cueRein_dir
        ]
    animals_of_interest_list = [
        LCDD_cueRein_animals,
        Gi_cueRein_animals,

        # LCDD_cueReinCNO_animals,
        # Gi_cueReinCNO_animals,

        # LCDD_cueReinPersist_animals,
        # Gi_cueReinPersist_animals,

        ALL_cueRein_animals
        ]

    groups_arrays = []
    data_set = []
    target = []
    epoch = [response_epoch[0], response_epoch[1]]
    for index in range(len(groups)):
        animals_of_interest = animals_of_interest_list[index]
        indir = indir_list[index]
        for animal in animals_of_interest:
            try:
                fovs = next(os.walk(os.path.join(indir, animal)))[1]
                for fov in sorted(fovs):
                    trials = np.load(os.path.join(indir, animal, fov, 'alignedActiveLeverCueData.npy'))
                    flags = np.load(os.path.join(indir, animal, fov, 'classificationFlags.npy'))
                    for neuron in range(trials.shape[2]):
                        data_set.append(trials[:,epoch[0]:epoch[1],neuron])
                        target.append(flags)
            except:
                print('*** ERROR: %s has no file ***'%(animal))

        arr = []
        for neuron in range(len(data_set)):
            #split test data
            X = np.array(data_set[neuron])
            y = np.array(target[neuron])

            try:
                X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, shuffle=False)

                #standardize data
                standardize = StandardScaler()
                X_train = standardize.fit_transform(X_train)
                X_test = standardize.fit_transform(X_test)

                #limit components
                reduce = PCA(n_components=3)
                X_train = reduce.fit_transform(X_train)
                X_test = reduce.fit_transform(X_test)
                
                # #fitting model
                # space = {
                #     'kernel': ['rbf'], 
                #     'gamma': [1e-2, 1e-1, 1e0, 1e1, 1e2],
                #     'C': [1e-2, 1e-1, 1e0, 1e1, 1e2]
                #     }

                #FIXME
                # define the function we want to minimise
                def objective(search_space):
                    model = RandomForestClassifier()
                    # model = LogisticRegression(**search_space)
                    model.fit(X_train, y_train)
                    y_pred = model.predict(X_test)
                    accuracy = accuracy_score(y_test, y_pred)
                    return {'loss': -accuracy, 'status': STATUS_OK}

                # define the values to search over for n_estimators
                # search_space = { 
                #     'gamma': hp.lognormal('gamma', 0, 1),
                #     'C': hp.randint('C', 1, 100)
                #     }
                search_space = {
                    'n_estimators': hp.randint('n_estimators', 0,3000),
                    'max_depth': hp.randint('max_depth', 0, 500),
                    }

                # set the hyperparam tuning algorithm
                algorithm=tpe.suggest

                best_params = fmin(
                    fn=objective,
                    space=search_space,
                    algo=algorithm,
                    max_evals=10)
                param_grid = space_eval(search_space, best_params)
                print(param_grid)

                clf = RandomForestClassifier(n_estimators=param_grid['n_estimators'], max_depth=param_grid['max_depth'])
                clf.fit(X_train, y_train)
            
                #probability of accurate classifications
                y_pred = clf.predict(X_test)
                y_pred_proba = clf.predict_proba(X_test)[:,1]

                # # classifier = LogisticRegression()
                # classifier = SVC()
                # model = RandomizedSearchCV(classifier, 
                #                            space, 
                #                            n_iter=5, 
                #                            cv=3, 
                #                            n_jobs=-1)
                # model.fit(X_train, y_train)

                # #probability of accurate classifications
                # y_pred = model.predict(X_test)
                # y_pred_proba = model.predict_proba(X_test)[:,1]

                accuracy = metrics.accuracy_score(y_test, y_pred)
                arr.append(accuracy)
            except Exception as e:
                print('***ERROR: %s ***'%(e))
        groups_arrays.append(arr)
    send_email_crash_notification('Decoding finished!')
except:
    log_stream = StringIO()
    logging.basicConfig(stream=log_stream, level=logging.INFO)
    logging.error("Exception occurred", exc_info=True)
    send_email_crash_notification('*** ERROR ***')
```
