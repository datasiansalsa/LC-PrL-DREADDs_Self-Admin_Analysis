### Algorithm

```python
#TRAINING DATA

#initialize group(s)
groups = ['CueRein', 'CueRein']
indir_list = [LCDD_cueRein_dir, Gi_cueRein_dir]
animals_of_interest_list = [LCDD_cueRein_animals, Gi_cueRein_animals]

print('========== STARTING ==========')

"""
current set up averages across neurons, resulting in a dataframe with average neuronal activity for each trial across windowsize
with a target variable as the last index
"""
print('...gathering data...')
train_set = []
train_target = []
for index in range(len(groups)):
    animals_of_interest = animals_of_interest_list[index]
    indir = indir_list[index]
    for animal in animals_of_interest:
        fovs = next(os.walk(os.path.join(indir, animal)))[1]
        for fov in sorted(fovs):
            trials = np.load(os.path.join(indir, animal, fov, 'alignedActiveLeverCueData.npy'))
            flags = np.load(os.path.join(indir, animal, fov, 'classificationFlags.npy'))
            for neuron in range(trials.shape[2]):
                for trial in range(trials.shape[0]):
                    train_set.append(trials[trial,:,neuron])
                    train_target.append(flags[trial])
#X and y training data
X = np.array(train_set)
y = np.array(train_target)

print('...standardizing data...')
#standardize data
standardize = OBJECT
stand_X = standardize.fit_transform(X)

print('...reducing data...')
#reduce dimensions while maintaining variance
reduction = DIMENSION_REDUCTION
reduced_X = reduction.fit_transform(stand_X)

print('...training data...')
# #accuracy
MODEL.fit(reduced_X, y)

print('========== FINISHED ==========')
```

```python
#SHUFFLE

groups = ['CueReinPersist', 'CueReinPersist']
indir_list = [LCDD_cueReinPersist_dir, Gi_cueReinPersist_dir]
animals_of_interest_list = [LCDD_cueReinPersist_animals, Gi_cueReinPersist_animals]

print('========== STARTING ==========')

shuffle_arr = []
for index in range(len(groups)):
    animals_of_interest = animals_of_interest_list[index]
    indir = indir_list[index]
    for animal in animals_of_interest:
        fovs = next(os.walk(os.path.join(indir, animal)))[1]
        for fov in sorted(fovs):
            trials = np.load(os.path.join(indir, animal, fov, 'alignedActiveLeverCueData.npy'))
            flags = np.load(os.path.join(indir, animal, fov, 'classificationFlags.npy'))

            num_trials = trials.shape[0]
            numneuronsinfov = trials.shape[2]
            if num_trials >= 6:
                positives = len(flags[flags==1])
                negatives = len(flags[flags==0])
                if positives < 3 or negatives < 3:
                    activeflag = np.hstack((np.zeros((int(len(flags)/2))), np.ones((int(len(flags)/2)))))
                else:
                    activeflag = np.hstack((np.zeros((negatives)), 
                                            np.ones((positives))))
                for neuron in range(numneuronsinfov):
                    y = activeflag 
                    X = trials[:,:,neuron]
                    X[np.where(np.isnan(X))] = np.nanmean(X)
                    stand_X_test = standardize.fit_transform(X)
                    reduction = DIMENSION_REDUCTION
                    reduced_X_test = reduction.fit_transform(stand_X_test)
                    yhat = MODEL.predict(reduced_X_test)
                    accuracy = accuracy_score(y, yhat)
                    shuffle_arr.append(accuracy)

#shuffle results
shuffled_shuffle_arr = np.random.permutation(shuffle_arr)

print('========== FINISHED ==========')
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import RocCurveDisplay

def plot_sklearn_roc_curve(y_real, y_pred, group, day, color):
    fpr, tpr, _ = roc_curve(y_real, y_pred)
    roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot(color=color)
    roc_display.figure_.set_size_inches(5,5)
    plt.plot([0, 1], [0, 1], color = 'k', linestyle='--')
    plt.legend([group])
    plt.title(day, fontsize=14, fontweight='bold')
```
