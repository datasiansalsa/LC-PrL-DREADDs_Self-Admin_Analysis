```python
#analysis by neuron

train_group = LCDD_cueRein_animals
train_dir = LCDD_cueRein_dir
epoch = [response_epoch[0], response_epoch[1]]
# epoch = [pre_window_size-int(3*7.5), pre_window_size] #Jim's baseline epoch
# epoch = [pre_window_size, pre_window_size+int(6*7.5)] #Jim's lever response epoch
base_neural_activity = []
base_target = []
for animal in train_group:
    try:
        fovs = next(os.walk(os.path.join(train_dir, animal)))[1]
        for fov in sorted(fovs):
            trials = np.load(os.path.join(train_dir, animal, fov, 'alignedActiveLeverCueData.npy'))
            flags = np.load(os.path.join(train_dir, animal, fov, 'classificationFlags.npy'))
            timeout_mean = np.nanmean(trials[np.where(flags==0),:,:])
            trials[np.where(flags==1),:,:] = trials[np.where(flags==1),:,:] - timeout_mean
            for neuron in range(trials.shape[2]):
                for trial in range(trials.shape[0]):
                    activity = trials[trial,epoch[0]:epoch[1],neuron]
                    base_neural_activity.append(activity)
                    base_target.append(flags[trial])
    except Exception as e:
        print('*** ERROR: %s ***'%(e))

#instantiate data
X = np.array(base_neural_activity)
y = np.array(base_target)

#split data
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75)

#pipeline
pipe = Pipeline([('scaler', MinMaxScaler(feature_range=(0,1))), 
                 ('classifier', SVC(C=.01, kernel='rbf'))])

#train data and predict accuracy
pipe.fit(X_train, y_train)
accuracy = pipe.score(X_test, y_test)
print('Trained accuracy: %s'%('{0:.3g}%'.format(accuracy*100)))  

LCDD_train_group = LCDD_cueReinPersist_animals
LCDD_train_dir = LCDD_cueReinPersist_dir
LCDD_neural_activity = []
LCDD_neuron_accuracy = []
for animal in LCDD_train_group:
    try:
        fovs = next(os.walk(os.path.join(LCDD_train_dir, animal)))[1]
        for fov in sorted(fovs):
            trials = np.load(os.path.join(LCDD_train_dir, animal, fov, 'alignedActiveLeverCueData.npy'))
            flags = np.load(os.path.join(LCDD_train_dir, animal, fov, 'classificationFlags.npy'))
            timeout_mean = np.nanmean(trials[np.where(flags==0),:,:])
            trials[np.where(flags==1),:,:] = trials[np.where(flags==1),:,:] - timeout_mean
            for neuron in range(trials.shape[2]):
                X = trials[:,epoch[0]:epoch[1],neuron]
                y = flags
                accuracy = pipe.score(X, y)
                LCDD_neuron_accuracy.append(accuracy)
    except Exception as e:
        print('*** ERROR: %s ***'%(e)) 

Gi_train_group = Gi_cueReinPersist_animals
Gi_train_dir = Gi_cueReinPersist_dir
Gi_neural_activity = []
Gi_neuron_accuracy = []
for animal in Gi_train_group:
    try:
        fovs = next(os.walk(os.path.join(Gi_train_dir, animal)))[1]
        for fov in sorted(fovs):
            trials = np.load(os.path.join(Gi_train_dir, animal, fov, 'alignedActiveLeverCueData.npy'))
            flags = np.load(os.path.join(Gi_train_dir, animal, fov, 'classificationFlags.npy'))
            timeout_mean = np.nanmean(trials[np.where(flags==0),:,:])
            trials[np.where(flags==1),:,:] = trials[np.where(flags==1),:,:] - timeout_mean
            for neuron in range(trials.shape[2]):
                X = trials[:,epoch[0]:epoch[1],neuron]
                y = flags
                accuracy = pipe.score(X, y)
                Gi_neuron_accuracy.append(accuracy)
    except Exception as e:
        print('*** ERROR: %s ***'%(e))

fig = plt.subplots(1,1, figsize=(10,4))
sns.set_style('darkgrid')
sns.kdeplot(LCDD_neuron_accuracy, cumulative=True, color='red', label='LCDD, Mean Classification Accuracy: %s'%('{0:.3g}%'.format(np.mean(LCDD_neuron_accuracy)*100)))
sns.ecdfplot(LCDD_neuron_accuracy, linewidth=.5, alpha =.7, color='red')
sns.kdeplot(Gi_neuron_accuracy, cumulative=True, color='k', label='DREADDs, Mean Classification Accuracy: %s'%('{0:.3g}%'.format(np.mean(Gi_neuron_accuracy)*100)))
sns.ecdfplot(Gi_neuron_accuracy, linewidth=.5, alpha =.7, color='k') 

plt.title('LCDD vs Gi-DREADDs Comparison', fontsize=12, fontweight='bold')
plt.legend(loc=2)
# confirm = input('Save figure? (y/n)')
# if confirm == 'y':
#     plt.savefig(os.path.join(iteration_results, 'LC-PrL-DREADDs_SA_%s.PDF'%(date)), format ='PDF')
#     print('Figure saved.')
# else:
#     print('Figure not saved.')
plt.text(.45,-.2, str(pipe))
plt.show()
```
